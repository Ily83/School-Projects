## Etape 2 - Etude de la convergence de la méthode ADAM 

```python
##########################################################
# Lecture des fichiers contenant les données des courbes #
##########################################################
Courbe_1 = pd.read_csv(r'C:\Users\tangu\Projet\Epoques\ADAM_50000.csv')
Courbe_1.columns = ['Epoques','Perte']
Courbe_2 = pd.read_csv(r'C:\Users\tangu\Projet\Epoques\ADAM_3000_LBFGS_47000.csv')
Courbe_2.columns = ['Epoques','Perte']
Courbe_3 = pd.read_csv(r'C:\Users\tangu\Projet\Epoques\ADAM_2500_LBFGS_47500.csv')
Courbe_3.columns = ['Epoques','Perte']
Courbe_4 = pd.read_csv(r'C:\Users\tangu\Projet\Epoques\ADAM_2000_LBFGS_48000.csv')
Courbe_4.columns = ['Epoques','Perte']
Courbe_5 = pd.read_csv(r'C:\Users\tangu\Projet\Epoques\ADAM_1500_LBFGS_48500.csv')
Courbe_5.columns = ['Epoques','Perte']
Courbe_6 = pd.read_csv(r'C:\Users\tangu\Projet\Epoques\ADAM_1000_LBFGS_49000.csv')
Courbe_6.columns = ['Epoques','Perte']
Courbe_7 = pd.read_csv(r'C:\Users\tangu\Projet\Epoques\ADAM_500_LBFGS_49500.csv')
Courbe_7.columns = ['Epoques','Perte']
```

```python
####################
# ADAM convergence #
####################
fig1 = plt.figure(figsize=(8,6))
plt.xlabel("Epoques")
plt.ylabel("Perte")
plt.title("Courbe d'apprentissage ADAM uniquement")
line1, = plt.plot(Courbe_1.iloc[0:3500,0],Courbe_1.iloc[0:3500,1:2].values)
plt.legend([line1],['ADAM'])
plt.yscale('log')
plt.show()
```

## Etape 3 - Etude de l'influence du nombre d'époques ADAM

```python
################
# Epoques ADAM #
################
fig2 = plt.figure(figsize=(8,6))
plt.xlabel("Epoques")
plt.ylabel("Perte")
plt.title("Influence de la fonction d'activation : courbes apprentissage")
line1, = plt.plot(Courbe_1.iloc[:,0],Courbe_1.iloc[:,1:2].values)
line2, = plt.plot(Courbe_2.iloc[:,0],Courbe_2.iloc[:,1:2].values)
line3, = plt.plot(Courbe_3.iloc[:,0],Courbe_3.iloc[:,1:2].values)
line4, = plt.plot(Courbe_4.iloc[:,0],Courbe_4.iloc[:,1:2].values)
line5, = plt.plot(Courbe_5.iloc[:,0],Courbe_5.iloc[:,1:2].values)
line6, = plt.plot(Courbe_6.iloc[:,0],Courbe_6.iloc[:,1:2].values)
line7, = plt.plot(Courbe_7.iloc[:,0],Courbe_7.iloc[:,1:2].values)
plt.legend([line1, line2, line3, line4, line5, line6, line7],['ADAM = 50 000','ADAM = 3000','ADAM = 2500','ADAM = 2000','ADAM = 1500','ADAM = 1000','ADAM = 500'])
plt.yscale('log')
plt.show()
```
