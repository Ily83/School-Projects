## Etape 2 - Comparaison des fonctions d'activations dites classiques

```python
##########################################################
# Lecture des fichiers contenant les données des courbes #
##########################################################
Courbe_1 = pd.read_csv(r'C:\Users\tangu\Projet\Fonction_Activation\ca_tanh.csv')
Courbe_1.columns = ['Epoques','Perte']
Courbe_2 = pd.read_csv(r'C:\Users\tangu\Projet\Fonction_Activation\ca_sine.csv')
Courbe_2.columns = ['Epoques','Perte']
Courbe_3 = pd.read_csv(r'C:\Users\tangu\Projet\Fonction_Activation\ca_sigmoid.csv')
Courbe_3.columns = ['Epoques','Perte']
Courbe_4 = pd.read_csv(r'C:\Users\tangu\Projet\Fonction_Activation\ca_swish.csv')
Courbe_4.columns = ['Epoques','Perte']
Courbe_5 = pd.read_csv(r'C:\Users\tangu\Projet\Fonction_Activation\ca_relu.csv')
Courbe_5.columns = ['Epoques','Perte']
```

```python
###########################
# Fonctions d'activations #
###########################
fig1 = plt.figure(figsize=(8,6))
plt.xlabel("Epoques")
plt.ylabel("Perte")
plt.title("Influence de la fonction d'activation : courbes apprentissage")
line1, = plt.plot(Courbe_1.iloc[:,0],Courbe_1.iloc[:,1:2].values)
line2, = plt.plot(Courbe_1.iloc[:,0],Courbe_2.iloc[:,1:2].values)
line3, = plt.plot(Courbe_1.iloc[:,0],Courbe_3.iloc[:,1:2].values)
line4, = plt.plot(Courbe_1.iloc[:,0],Courbe_4.iloc[:,1:2].values)
line5, = plt.plot(Courbe_1.iloc[:,0],Courbe_5.iloc[:,1:2].values)
plt.legend([line1, line2, line3, line4, line5],['TanH','Sine','Sigmoid','Swish','ReLU'])
plt.yscale('log')
plt.show()
```

## Etape 3 - Comparaison des fonctions d'activations découlantes de TanH

```python
##########################################################
# Lecture des fichiers contenant les données des courbes #
##########################################################
Courbe_1T = pd.read_csv(r'C:\Users\tangu\Projet\Fonction_Activation\ca_hardtanh.csv')
Courbe_1T.columns = ['Epoques','Perte']
Courbe_2T = pd.read_csv(r'C:\Users\tangu\Projet\Fonction_Activation\ca_tanhshrink.csv')
Courbe_2T.columns = ['Epoques','Perte']
```

```python
###########################
# Fonctions d'activations #
###########################
fig2 = plt.figure(figsize=(8,6))
plt.xlabel("Epoques")
plt.ylabel("Perte")
plt.title("Influence de la fonction d'activation : courbes apprentissage")
line1, = plt.plot(Courbe_1.iloc[:,0],Courbe_1.iloc[:,1:2].values)
line2, = plt.plot(Courbe_1T.iloc[:,0],Courbe_1T.iloc[:,1:2].values)
line3, = plt.plot(Courbe_2T.iloc[:,0],Courbe_2T.iloc[:,1:2].values)
plt.legend([line1, line2, line3],['TanH','HardTanH','TanhShrink'])
plt.yscale('log')
plt.show()
```
