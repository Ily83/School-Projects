## Etape 2 - Influence de la taille du dataset

```python
##########################################################
# Lecture des fichiers contenant les données des courbes #
##########################################################
Courbe_1 = pd.read_csv(r'C:\Users\tangu\Projet\Dataset\Taille\ca_ds_100.csv')
Courbe_1.columns = ['Epoques','Perte']
Courbe_2 = pd.read_csv(r'C:\Users\tangu\Projet\Dataset\Taille\ca_ds_1000.csv')
Courbe_2.columns = ['Epoques','Perte']
Courbe_3 = pd.read_csv(r'C:\Users\tangu\Projet\Dataset\Taille\ca_ds_1500.csv')
Courbe_3.columns = ['Epoques','Perte']
Courbe_4 = pd.read_csv(r'C:\Users\tangu\Projet\Dataset\Taille\ca_ds_2000.csv')
Courbe_4.columns = ['Epoques','Perte']
Courbe_5 = pd.read_csv(r'C:\Users\tangu\Projet\Dataset\Taille\ca_ds_2500.csv')
Courbe_5.columns = ['Epoques','Perte']
Courbe_6 = pd.read_csv(r'C:\Users\tangu\Projet\Dataset\Taille\ca_ds_4000.csv')
Courbe_6.columns = ['Epoques','Perte']
```

```python
##################
# Taille Dataset #
##################
fig1 = plt.figure(figsize=(8,6))
plt.xlabel("Epoques")
plt.ylabel("Perte")
plt.title("Influence de la taille du dataset : courbes apprentissage")
line1, = plt.plot(Courbe_1.iloc[:,0],Courbe_1.iloc[:,1:2].values)
line2, = plt.plot(Courbe_2.iloc[:,0],Courbe_2.iloc[:,1:2].values)
line3, = plt.plot(Courbe_3.iloc[:,0],Courbe_3.iloc[:,1:2].values)
line4, = plt.plot(Courbe_4.iloc[:,0],Courbe_4.iloc[:,1:2].values)
line5, = plt.plot(Courbe_5.iloc[:,0],Courbe_5.iloc[:,1:2].values)
line6, = plt.plot(Courbe_6.iloc[:,0],Courbe_6.iloc[:,1:2].values)

plt.legend([line1, line2, line3, line4, line5, line6],['N = 100','N = 1000','N = 1500','N = 2000', 'N = 2500', 'N = 4000'])
plt.yscale('log')
plt.show()
```

```python
###########################
# Perte en fonction de lr #
###########################
# Liste des dataset
Learning_Rate=[100,1000,1500,2000,2500,4000]
# On fait une moyenne de la fonction de perte sur les 1000 dernières itérations
Perte_1 = sum(Courbe_1.iloc[6000:7000,1:2].values)/1000
Perte_2 = sum(Courbe_2.iloc[6000:7000,1:2].values)/1000
Perte_3 = sum(Courbe_3.iloc[6000:7000,1:2].values)/1000
Perte_4 = sum(Courbe_4.iloc[6000:7000,1:2].values)/1000
Perte_5 = sum(Courbe_5.iloc[6000:7000,1:2].values)/1000
Perte_6 = sum(Courbe_6.iloc[6000:7000,1:2].values)/1000

# Liste des valeurs de la fonction de perte
Perte = [Perte_1,Perte_2,Perte_3,Perte_4,Perte_5,Perte_6]
# On trace la figure
fig2 = plt.figure(figsize=(8,6))
plt.xlabel("Taux apprentissage")
plt.ylabel("Perte")
plt.title("Evolution de la fonction de perte en fonction du taux apprentissage")
line1, = plt.plot(Learning_Rate,Perte,'-x')
plt.xscale('log')
plt.show()
```
